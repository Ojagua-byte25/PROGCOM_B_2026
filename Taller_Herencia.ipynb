```python

class Vehiculo:
    def __init__(self, marca):
        self.marca = marca

    def mostrar_marca(self):
        print("Marca:", self.marca)

class Carro(Vehiculo):
    def __init__(self, marca, modelo):
        super().__init__(marca)
        self.modelo = modelo

mi_carro = Carro("Lamborghini", "350 GT")
mi_carro.mostrar_marca()
print("Modelo:", mi_carro.modelo)
```

```python
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

class Estudiante(Persona):
    def __init__(self, nombre, curso):
        super().__init__(nombre)
        self.curso = curso

mi_estudiante = Estudiante("José", "Derecho")
print("Nombre:", mi_estudiante.nombre)
print("Curso:", mi_estudiante.curso)
```

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

class Perro(Animal):
    def __init__(self, nombre, raza):
        super().__init__(nombre)
        self.raza = raza

mi_perro = Perro("Luigi", "Rottweiler")
print("Nombre:", mi_perro.nombre)
print("Raza:", mi_perro.raza)
```

```python
class Ropa:
    def __init__(self, color):
        self.color = color

class Camisa(Ropa):
    def __init__(self, color, talla):
        super().__init__(color)
        self.talla = talla

mi_camisa = Camisa("Rojo", "XL")
print("Color:", mi_camisa.color)
print("Talla:", mi_camisa.talla)
```

```python
class Dispositivo:
    def __init__(self, marca):
        self.marca = marca

class Celular(Dispositivo):
    def __init__(self, marca, memoria):
        super().__init__(marca)
        self.memoria = memoria

mi_celular = Celular("Xiaomi", "256GB")
print("Marca:", mi_celular.marca)
print("Memoria:", mi_celular.memoria)
```

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print("Hola, me llamo", self.nombre)

class Profesor(Persona):
    def __init__(self, nombre, edad, materia):
        super().__init__(nombre, edad)
        self.materia = materia

mi_profesor = Profesor("Ricardo", 38, "Biología")
mi_profesor.saludar()
print("Edad:", mi_profesor.edad)
print("Materia:", mi_profesor.materia)
```

```python
class Cuenta:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def mostrar_saldo(self):
        print("Saldo:", self.saldo)

class CuentaCorriente(Cuenta):
    def __init__(self, titular, saldo, numero_cuenta):
        super().__init__(titular, saldo)
        self.numero_cuenta = numero_cuenta

mi_cuenta = CuentaCorriente("Luz García", 6000, "002-4567")
print("Titular:", mi_cuenta.titular)
print("Número de cuenta:", mi_cuenta.numero_cuenta)
mi_cuenta.mostrar_saldo()
```

```python
class Empleado:
    def __init__(self, nombre, salario):
        self.nombre = nombre
        self.salario = salario

    def trabajar(self):
        print(self.nombre, "está trabajando")

class Disenador(Empleado):
    def __init__(self, nombre, salario, programa):
        super().__init__(nombre, salario)
        self.programa = programa

mi_disenador = Disenador("Jesicca", 5000, "Edición")
print("Nombre:", mi_disenador.nombre)
print("Salario:", mi_disenador.salario)
print("Programa que utiliza:", mi_disenador.programa)
```

```python
class Figura:
    def mostrar_figura(self):
        print("Tipo de figura: Triángulo")

class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        area = (self.base * self.altura) / 2
        print("Área del triángulo:", area)

mi_triangulo = Triangulo(5, 8)
mi_triangulo.mostrar_figura()
mi_triangulo.calcular_area()
```

```python
class Deportista:
    def __init__(self, nombre, deporte):
        self.nombre = nombre
        self.deporte = deporte

    def entrenar(self):
        print(self.nombre, "está entrenando")

class Nadador(Deportista):
    def __init__(self, nombre, deporte, estilo):
        super().__init__(nombre, deporte)
        self.estilo = estilo

mi_nadador = Nadador("Richard", "Natación", "Braza")
print("Nombre:", mi_nadador.nombre)
print("Deporte:", mi_nadador.deporte)
print("Estilo de natación:", mi_nadador.estilo)
```
