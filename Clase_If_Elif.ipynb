#IF/ELIF
Se comporta como un switch, ES DECIR, ES PARA MULTIPLE SLECCION
SINTAXIS


```
if (C1): aquí se ejecuta código si C1 es TRUE
elif (C2)
```




````python
if edad<10: print("Aún eres un niño")
elif edad>=10 and edad<14: print("Eres un preadolecente")
elif edad>=14 and edad<18: print("Eres un adolescente")
elif edad>=18 and edad<30: print("Eres un adulto joven")
else print("Eres un adulto")
````
