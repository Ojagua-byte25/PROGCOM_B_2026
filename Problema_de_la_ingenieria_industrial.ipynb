```python
# ===============================================
#  REGISTRO DE ACCIDENTES LABORALES
#  Ingenieria Industrial - SST
# ===============================================

accidentes = []
gravedad_total = 0

print("=== REGISTRO DE ACCIDENTES LABORALES ===")

n = int(input("¿Cuantos accidentes vas a registrar? "))

for i in range(n):
    print(f"\n--- Accidente #{i+1} ---")
    trabajador = input("Nombre del trabajador: ")
    area = input("Area de trabajo: ")
    gravedad = int(input("Nivel de gravedad (1=Leve, 2=Moderado, 3=Grave): "))

    accidentes.append([trabajador, area, gravedad])
    gravedad_total += gravedad

# Mostrar resultados
print("\n========== REPORTE FINAL ==========")
print(f"{'#':<4} {'Trabajador':<20} {'Area':<15} {'Gravedad'}")
print("-" * 50)

for i in range(len(accidentes)):
    a = accidentes[i]
    if a[2] == 1:
        nivel = "Leve"
    elif a[2] == 2:
        nivel = "Moderado"
    else:
        nivel = "GRAVE"
    print(f"{i+1:<4} {a[0]:<20} {a[1]:<15} {nivel}")

promedio = gravedad_total / n
print(f"\nPromedio de gravedad: {promedio:.2f}")

if promedio >= 2.5:
    print("⚠ ALERTA: Nivel de riesgo alto en la empresa.")
else:
    print("✔ Nivel de riesgo aceptable.")
```
