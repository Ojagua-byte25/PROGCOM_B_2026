#WHILE
1. Cuando no se sabe cuántas veces se repite una acción.
2. Es gobernado por una condición de parada.
3. Es necesaria una condición inicial, la cual **DEBE** cambiar adentro del ciclo.
Sintaxis
```
While(condicion):
codigo a repetir
modificador de condición
```



````python
# Suponer que la cantidad maxima de votantes son 10
# Suponer que son 5 candidatos
v1=0
v2=0
v3=0
v4=0
v5=0
vn=0
vb=0
persona=0
votico="s"
documentos_votados = []  # Almacena los IDs que ya votaron

while votico=="s" and persona<=10:
    id=input("Ingresa tu numero de documento: ")

    if id in documentos_votados:
        print(f"El documento {id} ya emitió su voto anteriormente.")
    else:
        documentos_votados.add(id)
        voto=input("Ingrese por quien quiere votar: \n1. Elefante blanco \n2. Pantera rosa \n3. Jotape Hernandez \n4. Paloma \n5. Quevedito \n6. Voto en blanco \n")
        if voto=="1": v1+=1
        elif voto=="2": v2+=1
        elif voto=="3": v3+=1
        elif voto=="4": v4+=1
        elif voto=="5": v5+=1
        elif voto=="6": vb+=1
        else: vn+=1
        print("Gracias por su voto")
        persona+=1

    votico=input("Hay personas antes de las 4:00PM? (s/n): ")
    print("")

print(f" RESULTADOS: \n1 Elefante blanco: {v1} \n2. Pantera rosa: {v2} \n3. Jotape Hernandez: {v3} \n4. Paloma: {v4} \n5. Quevedito: {v5} \n6. Voto en blanco {vb} \nVoto nulo: {vn}")
````
