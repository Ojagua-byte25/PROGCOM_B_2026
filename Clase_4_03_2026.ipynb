#WHILE

Sintaxis

```
While (CI):
  lo que se va a repetir modificador de CI
```
**OJO!!**
- While arranca si CI es VERDADERA
- Modificadores de CI:pregunta, contador, acumulador, cambio de variable

````python
c1=0
c2=0
c3=0
c4=0
nulo=0
votos="s"
while votos=="s":
  voto=int(input("Ingrese el numero de su candidato (1-3),\n0 para voto en blanco."))
if voto==1: c1=c1+1
elif voto==2: c2+=1 #c2=c2+1
elif voto==3: c3+=1
elif voto==4: c4+=1
else: nulo+=1
print("Hay más votos? s para si, n para no")
votos=votos[0].lower()
print("")
print(f"Total de votos:\nCandidato 1={c1}\nCandidato 2={c2}\nCandidato 3={c3}?\nVoto en blanco={c4}\nNulo={nulo}")
````
