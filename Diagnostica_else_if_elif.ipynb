```python
print("=== ECUACIÓN CUADRÁTICA ===")
a = float(input("a = "))
b = float(input("b = "))
c = float(input("c = "))

if a == 0:
    print("No es ecuación cuadrática")
else:
    d = b**2 - 4*a*c  # discriminante

    if d < 0:
        print("No tiene solución real")
    elif d == 0:
        x = -b / (2*a)
        print("Solución única: x =", x)
    else:
        x1 = (-b + d**0.5) / (2*a)
        x2 = (-b - d**0.5) / (2*a)
        print("Soluciones: x1 =", x1, "y x2 =", x2)
```

```python

print("=== PAR O IMPAR ===")
numero = int(input("Escribe un número entero: "))

if numero % 2 == 0:
    print(f"El número {numero} es PAR")
else:
    print(f"El número {numero} es IMPAR")
print(f"{numero} dividido entre 2:")
print(f"Residuo = {numero % 2}")
if numero % 2 == 0:
    print("Como el residuo es 0  el número es PAR")
else:
    print("Como el residuo NO es 0  el número es IMPAR")
```
