Escribe un programa que calcule la suma de números naturales hasta n.

```python
#FOR
suma = 0
num=int(input("Ingrese el número: "))
for i in range(1, num + 1):
    suma+=i
print("La suma de los números naturales es:",suma)
```

```python
#WHILE
suma=0
num=int(input("Ingrese un número: "))
i = 1
while i <= num:
    suma+=i
    i=i+1
print("La suma de los números naturales e:", suma)
```

Desarrolla un programa que calcule el factorial de un número. Recuerda que el factorial de 0 es 1 y no existen factoriales negativos.

```python
#FOR
num=int(input("Ingrese un número: "))
if num < 0:
  print("No existen factoriales negativos")
else:
  factorial=1
  for i in range(1, num + 1):
      factorial=factorial*i
  print("El factorial es:", factorial)
```

```python
#WHILE
num = int(input("Ingrese un número: "))
if num<0:
  print("No existen factoriales negativos")
else:
  factorial=1
  i=1
  while i <= num:
      factorial=factorial*i
      i=i+1
  print("El factorial es:", factorial)
```

Escribe un programa que adivine entre un conjunto de diez colores definidos por el usuario, con un máximo de 10 intentos.

```python
#FOR
colores=[]
for i in range(10):
  color=input("Escribe un color: ")
  colores.append(color)
secreto = colores[0]
for i in range(10):
  intento=input("Adivina el color: ")
  if intento==secreto:
      print("¡Ganaste!")
      break
  else:
      print("No es ese color")
print("El color era:", secreto)
```

```python
#WHILE
colores=[]
i=0
while i < 10:
  color = input("Escribe un color: ")
  colores.append(color)
  i=i+1
secreto = colores[0]
intentos = 0
while intentos < 10:
  intento = input("Adivina el color: ")
  if intento == secreto:
      print("¡Ganaste!")
      break
  else:
      print("No es ese color")
  intentos = intentos + 1
print("El color era:", secreto)
```
