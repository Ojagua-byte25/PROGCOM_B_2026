#FUNCIONES
Es un pedazo de codigo "muerto" que no se ejecuta hasta que el codigo principal lo ejecute.

Ventanas
1. Organiza el codigo
2. Habilita un nuevo paradigma de programación: programación modular
3. Habilita el copiado dinamico (reutilización del código)

Sintaxis
```
def nombre_funcion(p1,p2,p3,...,pn)
 aqui va el código
 return (dato)
```
Donde:
* p1,p2,...pn son parametros
* return retorna un dato

Existen 4 tipos diferentes de funciones
1. Sin parametro, sin retorno
2. Sin parametro, con retorno
3. Con parametro, sin retorno
4. Con parametro, con retorno


````python
#Con parámetro, sin retorno
def cuadrado(a):
  print(f"a*2={a*a}")
````

````python
#Con parámetro, con retorno
def cuadrado_venganza(a):
  return a*a
````
