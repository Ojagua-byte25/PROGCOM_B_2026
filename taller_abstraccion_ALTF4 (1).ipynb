```python
# Ejercicio 1: El Cajero Automático

class CajeroAutomatico:

    def __init__(self, saldo):
        self.saldo = saldo

    def __autenticar_sesion(self):
        print("  [Oculto] Verificando sesión con el servidor del banco...")

    def __consultar_base_de_datos(self):
        print("  [Oculto] Consultando base de datos para traer saldo actualizado...")

    def __verificar_fondos(self, monto):
        print("  [Oculto] Verificando que haya fondos suficientes...")
        return self.saldo >= monto

    def __registrar_transaccion(self, tipo, monto):
        print("  [Oculto] Registrando transacción en el historial del banco...")

    def __validar_cuenta_destino(self, cuenta):
        print("  [Oculto] Validando que la cuenta destino exista...")

    def __cifrar_datos(self):
        print("  [Oculto] Cifrando datos para envío seguro por la red...")

    def consultar_saldo(self):
        self.__autenticar_sesion()
        self.__consultar_base_de_datos()
        print("Tu saldo disponible es:", self.saldo)

    def retirar_dinero(self, monto):
        hay_fondos = self.__verificar_fondos(monto)
        if hay_fondos:
            self.saldo = self.saldo - monto
            self.__registrar_transaccion("retiro", monto)
            print("Retiro exitoso de:", monto)
        else:
            print("Fondos insuficientes")

    def transferir(self, monto, cuenta_destino):
        self.__validar_cuenta_destino(cuenta_destino)
        self.__cifrar_datos()
        hay_fondos = self.__verificar_fondos(monto)
        if hay_fondos:
            self.saldo = self.saldo - monto
            self.__registrar_transaccion("transferencia", monto)
            print("Transferencia exitosa de", monto, "a la cuenta", cuenta_destino)
        else:
            print("Fondos insuficientes")


cajero = CajeroAutomatico(saldo=500000)

print("--- Consultar saldo ---")
cajero.consultar_saldo()

print("\n--- Retirar dinero ---")
cajero.retirar_dinero(100000)

print("\n--- Transferir ---")
cajero.transferir(50000, "123456789")
```

```python
# Ejercicio 2: Personaje de Videojuego

class Personaje:

    def __init__(self, energia):
        self.energia = energia

    def __calcular_dano_armadura(self, armadura_enemigo):
        dano_base = 50
        dano_final = dano_base - armadura_enemigo
        return dano_final

    def __calcular_dano_energia(self):
        if self.energia < 30:
            return 0.5
        else:
            return 1.0

    def atacar(self, armadura_enemigo):
        dano = self.__calcular_dano_armadura(armadura_enemigo)
        multiplicador = self.__calcular_dano_energia()
        dano_total = dano * multiplicador
        print("El personaje ataca! Daño infligido:", dano_total)


print("--- Personaje con mucha energía ---")
heroe1 = Personaje(energia=80)
heroe1.atacar(armadura_enemigo=10)

print("\n--- Personaje con poca energía ---")
heroe2 = Personaje(energia=20)
heroe2.atacar(armadura_enemigo=10)
```

```python
# Ejercicio 3: Sistema de Envíos

class SistemaDeEnvios:

    def __enviar_por_carretera(self, destino, paquete):
        print("  [Oculto] Asignando camión disponible para la ruta...")
        print("  [Oculto] Calculando ruta terrestre óptima...")
        print("  Paquete", paquete, "enviado a", destino, "en camión")

    def __enviar_por_barco(self, destino, paquete):
        print("  [Oculto] Reservando espacio en el barco de carga...")
        print("  [Oculto] Calculando ruta marítima y tiempo de llegada...")
        print("  Paquete", paquete, "enviado a", destino, "en barco")

    def __enviar_por_drone(self, destino, paquete):
        print("  [Oculto] Calculando batería necesaria para el vuelo...")
        print("  [Oculto] Verificando espacio aéreo disponible...")
        print("  Paquete", paquete, "enviado a", destino, "en drone")

    def enviar_paquete(self, destino, paquete, peso_kg):
        print("Procesando envío de", paquete, "a", destino)
        if peso_kg <= 1:
            self.__enviar_por_drone(destino, paquete)
        elif peso_kg <= 100:
            self.__enviar_por_carretera(destino, paquete)
        else:
            self.__enviar_por_barco(destino, paquete)


sistema = SistemaDeEnvios()

print("--- Paquete pequeño (drone) ---")
sistema.enviar_paquete("Bucaramanga", "Sobre de documentos", peso_kg=0.5)

print("\n--- Paquete mediano (camión) ---")
sistema.enviar_paquete("Bogotá", "Nevera", peso_kg=60)

print("\n--- Paquete pesado (barco) ---")
sistema.enviar_paquete("Miami", "Maquinaria industrial", peso_kg=500)
```
