```python
# Ejercicio 1: Animales

class Animal:
    def sonido(self):
        print("El animal hace un sonido")

class Perro(Animal):
    def sonido(self):
        print("El perro dice: Guau!")

class Gato(Animal):
    def sonido(self):
        print("El gato dice: Miau!")

class Vaca(Animal):
    def sonido(self):
        print("La vaca dice: Muuu!")

perro = Perro()
gato = Gato()
vaca = Vaca()

perro.sonido()
gato.sonido()
vaca.sonido()
```

```python
# Ejercicio 2: Vehículos

class Vehiculo:
    def mover(self):
        print("El vehículo se mueve")

class Carro(Vehiculo):
    def mover(self):
        print("El carro se desplaza por la carretera")

class Avion(Vehiculo):
    def mover(self):
        print("El avión vuela por el aire")

class Barco(Vehiculo):
    def mover(self):
        print("El barco navega por el agua")

carro = Carro()
avion = Avion()
barco = Barco()

carro.mover()
avion.mover()
barco.mover()
```

```python
# Ejercicio 3: Profesiones

class Empleado:
    def trabajar(self):
        print("El empleado está trabajando")

class Profesor(Empleado):
    def trabajar(self):
        print("El profesor está dando clases a sus estudiantes")

class Ingeniero(Empleado):
    def trabajar(self):
        print("El ingeniero está diseñando y construyendo proyectos")

class Doctor(Empleado):
    def trabajar(self):
        print("El doctor está atendiendo pacientes")

profesor = Profesor()
ingeniero = Ingeniero()
doctor = Doctor()

profesor.trabajar()
ingeniero.trabajar()
doctor.trabajar()
```

```python
# Ejercicio 5: Deportes

class Deporte:
    def jugar(self):
        print("Se está practicando un deporte")

class Futbol(Deporte):
    def jugar(self):
        print("En el fútbol se juega pateando un balón hacia el arco contrario")

class Baloncesto(Deporte):
    def jugar(self):
        print("En el baloncesto se juega lanzando el balón a la canasta")

class Tenis(Deporte):
    def jugar(self):
        print("En el tenis se juega golpeando la pelota con una raqueta")

futbol = Futbol()
baloncesto = Baloncesto()
tenis = Tenis()

futbol.jugar()
baloncesto.jugar()
tenis.jugar()
```

```python
# Ejercicio 6: Instrumentos musicales

class Instrumento:
    def tocar(self):
        print("Se está tocando un instrumento")

class Guitarra(Instrumento):
    def tocar(self):
        print("La guitarra suena al rasgar sus cuerdas: ¡Tararán!")

class Piano(Instrumento):
    def tocar(self):
        print("El piano suena al presionar sus teclas: ¡Tilín tilín!")

class Bateria(Instrumento):
    def tocar(self):
        print("La batería suena al golpear los tambores: ¡Pum pum pum!")

guitarra = Guitarra()
piano = Piano()
bateria = Bateria()

guitarra.tocar()
piano.tocar()
bateria.tocar()
```

```python
# Ejercicio 7: Medios de transporte

class Transporte:
    def viajar(self):
        print("El transporte está viajando")

class Bus(Transporte):
    def viajar(self):
        print("El bus viaja por la carretera llevando muchos pasajeros")

class Moto(Transporte):
    def viajar(self):
        print("La moto viaja rápido por las calles de la ciudad")

class Bicicleta(Transporte):
    def viajar(self):
        print("La bicicleta viaja pedaleando por el ciclovía")

bus = Bus()
moto = Moto()
bicicleta = Bicicleta()

bus.viajar()
moto.viajar()
bicicleta.viajar()
```

```python
# Ejercicio 8: Videojuegos

class Personaje:
    def atacar(self):
        print("El personaje está atacando")

class Mago(Personaje):
    def atacar(self):
        print("El mago ataca lanzando una bola de fuego mágica")

class Arquero(Personaje):
    def atacar(self):
        print("El arquero ataca disparando una flecha con su arco")

class Guerrero(Personaje):
    def atacar(self):
        print("El guerrero ataca golpeando con su espada")

mago = Mago()
arquero = Arquero()
guerrero = Guerrero()

mago.atacar()
arquero.atacar()
guerrero.atacar()
```

```python
# Ejercicio 9: Calculadora

class Calculadora:
    def operar(self, numero1, numero2, operador):
        if operador == "+":
            resultado = numero1 + numero2
            print(numero1, "+", numero2, "=", resultado)
        elif operador == "-":
            resultado = numero1 - numero2
            print(numero1, "-", numero2, "=", resultado)
        else:
            print("Operador no válido")

calc = Calculadora()

calc.operar(10, 5, "+")
calc.operar(20, 8, "-")
calc.operar(100, 50, "+")
calc.operar(30, 12, "-")
```

```python
# Ejercicio 10: Conversión de texto

class Texto:
    def convertir(self, texto, formato):
        if formato == "mayus":
            print("Texto en mayúsculas:", texto.upper())
        elif formato == "minus":
            print("Texto en minúsculas:", texto.lower())
        else:
            print("Formato no válido")

t = Texto()

t.convertir("Hola Mundo", "mayus")
t.convertir("Hola Mundo", "minus")
t.convertir("Python es genial", "mayus")
t.convertir("PROGRAMACION ORIENTADA A OBJETOS", "minus")
```
