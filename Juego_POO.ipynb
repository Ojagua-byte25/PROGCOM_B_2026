```python
"""
====================================================
  FÁBRICA DE HÉROES — Juego didáctico de POO
  Tema: Clases y Objetos en Python
====================================================
"""
import time
import random

# ─────────────────────────────────────────
#  COLORES (ANSI) — funciona en terminal
# ─────────────────────────────────────────
class Color:
    MORADO  = "\033[95m"
    AZUL    = "\033[94m"
    VERDE   = "\033[92m"
    AMARILLO= "\033[93m"
    ROJO    = "\033[91m"
    NEGRITA = "\033[1m"
    RESET   = "\033[0m"

def c(texto, color):
    return f"{color}{texto}{Color.RESET}"

def separador():
    print(c("─" * 50, Color.AZUL))

def titulo(texto):
    separador()
    print(c(f"  {texto}", Color.NEGRITA))
    separador()

def pausa():
    input(c("\n  Presiona Enter para continuar...", Color.AMARILLO))

# ─────────────────────────────────────────
#  CLASE DEL JUEGO — ¡esto ya ES un ejemplo de POO!
# ─────────────────────────────────────────
class JuegoPOO:
    """
    Esta clase representa el juego completo.
    ¡Fíjate! El juego en sí mismo ES un objeto de POO.
    Atributos: score, heroe_nombre, heroe_tipo, heroe_vida, heroe_fuerza
    Métodos: cada función es un método del juego
    """

    def __init__(self):
        # ATRIBUTOS del objeto juego
        self.score = 0
        self.heroe_nombre = ""
        self.heroe_tipo = ""
        self.heroe_vida = 0
        self.heroe_fuerza = 0
        self.total_rondas = 5
        self.ronda_actual = 0

    # ── MÉTODOS AUXILIARES ──────────────────

    def mostrar_progreso(self):
        bloques = int((self.ronda_actual / self.total_rondas) * 20)
        barra = "█" * bloques + "░" * (20 - bloques)
        print(c(f"\n  Progreso: [{barra}] Ronda {self.ronda_actual}/{self.total_rondas}", Color.AZUL))
        print(c(f"  Puntaje actual: {self.score} pts\n", Color.AMARILLO))

    def opcion_multiple(self, pregunta, opciones, correcta):
        """
        Muestra una pregunta de opción múltiple.
        Retorna True si acertó, False si no.
        """
        print(c(f"\n  {pregunta}", Color.NEGRITA))
        print()
        letras = ["a", "b", "c", "d"]
        for i, op in enumerate(opciones):
            print(f"    {c(letras[i]+')', Color.MORADO)} {op}")

        print()
        while True:
            resp = input(c("  Tu respuesta (a/b/c/d): ", Color.AZUL)).strip().lower()
            if resp in letras[:len(opciones)]:
                break
            print(c("  Ingresa solo a, b, c o d.", Color.ROJO))

        idx = letras.index(resp)
        if idx == correcta:
            print(c("\n  ✓ ¡CORRECTO!", Color.VERDE))
            return True
        else:
            print(c(f"\n  ✗ Incorrecto. La respuesta era: {letras[correcta]}) {opciones[correcta]}", Color.ROJO))
            return False

    # ── PANTALLAS ───────────────────────────

    def intro(self):
        print()
        titulo("FÁBRICA DE HÉROES — POO Didáctico")
        print("""
  Aprende Clases y Objetos construyendo héroes épicos.

  En este juego vas a aprender:
""")
        print(c("  · Qué es una CLASE", Color.MORADO))
        print(c("  · Qué es un OBJETO (instancia)", Color.VERDE))
        print(c("  · Diferencia entre ATRIBUTOS y MÉTODOS", Color.AZUL))
        print(c("  · Cómo instanciar una clase en Python", Color.AMARILLO))
        print(f"""
  Hay {self.total_rondas} rondas. Cada una vale hasta 20 pts.
  ¡Máximo posible: 100 puntos!
""")
        pausa()

    # ── RONDA 1: Clase vs Objeto ─────────────

    def ronda1_concepto(self):
        self.ronda_actual = 1
        titulo("RONDA 1 — ¿Clase o Instancia?")
        self.mostrar_progreso()

        print(c("  TEORÍA:", Color.AMARILLO))
        print("""
  Una CLASE es el molde o plantilla.
  Un OBJETO es una copia creada a partir de ese molde.

  Ejemplo del mundo real:
  ┌─────────────────────────────────────────┐
  │  CLASE  →  el plano de una casa         │
  │  OBJETO →  la casa construida con él    │
  └─────────────────────────────────────────┘

  Puedes construir MUCHOS objetos desde UNA sola clase.
""")

        correcta = self.opcion_multiple(
            "¿Cuál de estos describe mejor una CLASE en POO?",
            [
                "La cookie de avena recién horneada",
                "El plano del arquitecto para construir casas",
                "La lista de ingredientes de una receta específica",
                "Una foto del edificio terminado",
            ],
            correcta=1
        )

        if correcta:
            self.score += 20
            print(c("  +20 puntos — ¡La clase es el plano, el objeto es el edificio!", Color.VERDE))
        else:
            print(c("  La clase define la ESTRUCTURA. Los objetos son las instancias concretas.", Color.AMARILLO))

        pausa()

    # ── RONDA 2: Atributos vs Métodos ────────

    def ronda2_atributos_metodos(self):
        self.ronda_actual = 2
        titulo("RONDA 2 — Atributos vs Métodos")
        self.mostrar_progreso()

        print(c("  TEORÍA:", Color.AMARILLO))
        print("""
  ATRIBUTOS = características / datos del objeto
              (nombre, vida, fuerza, color...)

  MÉTODOS   = acciones que puede hacer el objeto
              (atacar(), curar(), correr()...)

  En Python se ven así:

    class Heroe:
        def __init__(self, nombre, vida):
            self.nombre = nombre    ← ATRIBUTO
            self.vida   = vida      ← ATRIBUTO

        def atacar(self):           ← MÉTODO
            return "¡Ataca!"
""")

        elementos = [
            ("nombre",    "atributo"),
            ("atacar()",  "metodo"),
            ("vida",      "atributo"),
            ("curar()",   "metodo"),
            ("fuerza",    "atributo"),
            ("correr()",  "metodo"),
        ]
        random.shuffle(elementos)

        print(c("  EJERCICIO: clasifica cada elemento (a = atributo, m = método)\n", Color.NEGRITA))
        correctas = 0
        for elem, tipo in elementos:
            while True:
                resp = input(c(f"    {elem:12s} → ", Color.MORADO) + "(a/m): ").strip().lower()
                if resp in ("a", "m"):
                    break
                print(c("      Escribe 'a' para atributo o 'm' para método.", Color.ROJO))

            correcto = (resp == "a" and tipo == "atributo") or (resp == "m" and tipo == "metodo")
            if correcto:
                print(c(f"      ✓ Correcto — {elem} es un {tipo}", Color.VERDE))
                correctas += 1
            else:
                print(c(f"      ✗ Era un {tipo}", Color.ROJO))

        pts = round((correctas / len(elementos)) * 20)
        self.score += pts
        print(c(f"\n  Resultado: {correctas}/{len(elementos)} correctas → +{pts} puntos", Color.AMARILLO))
        pausa()

    # ── RONDA 3: Leer código ─────────────────

    def ronda3_leer_codigo(self):
        self.ronda_actual = 3
        titulo("RONDA 3 — Lee el código")
        self.mostrar_progreso()

        print(c("  Analiza esta clase Python:", Color.AMARILLO))
        print("""
    class Heroe:
        def __init__(self, nombre, vida, fuerza):
            self.nombre  = nombre
            self.vida    = vida
            self.fuerza  = fuerza

        def atacar(self):
            return self.nombre + " ataca!"

        def curar(self, puntos):
            self.vida += puntos

        def huir(self):
            return self.nombre + " huye!"
""")

        preguntas = [
            (
                "¿Cuántos ATRIBUTOS tiene esta clase?",
                ["1 (nombre)", "2 (atacar, curar)", "3 (nombre, vida, fuerza)", "4 (incluyendo self)"],
                2,
                "nombre, vida y fuerza son los 3 atributos definidos con self.x = x"
            ),
            (
                "¿Cuántos MÉTODOS tiene esta clase?",
                ["2 (atacar y curar)", "3 (atacar, curar, huir)", "4 (incluyendo __init__)", "1 (solo __init__)"],
                2,
                "Los métodos son: atacar(), curar() y huir(). __init__ también es un método, ¡pero la respuesta 3 contaría 4!"
            ),
        ]

        pts_ronda = 0
        for pregunta, ops, correcta, explicacion in preguntas:
            ok = self.opcion_multiple(pregunta, ops, correcta)
            if ok:
                pts_ronda += 10
            else:
                print(c(f"  Pista: {explicacion}", Color.AZUL))

        self.score += pts_ronda
        print(c(f"\n  +{pts_ronda} puntos esta ronda", Color.AMARILLO))
        pausa()

    # ── RONDA 4: Crear héroe (instanciar) ────

    def ronda4_instanciar(self):
        self.ronda_actual = 4
        titulo("RONDA 4 — ¡Instancia tu héroe!")
        self.mostrar_progreso()

        print(c("  TEORÍA:", Color.AMARILLO))
        print("""
  Instanciar = crear un OBJETO desde una CLASE.

  Así se hace en Python:

    mi_heroe = Heroe("Shadowbyte", 100, 85)
    #           ↑ clase   ↑ argumentos del __init__

  Cada objeto tiene sus PROPIOS valores de atributos.
  Puedes crear todos los que quieras:

    heroe1 = Heroe("Mateo", 100, 90)
    heroe2 = Heroe("Ana",    80, 95)
    heroe3 = Heroe("Zarkon", 60, 99)
""")
        print(c("  Ahora vas a crear TU héroe:\n", Color.NEGRITA))

        # Pedir datos
        while True:
            nombre = input(c("  Nombre de tu héroe: ", Color.MORADO)).strip()
            if nombre:
                break
            print(c("  El nombre no puede estar vacío.", Color.ROJO))

        tipos = {"1": "Mago", "2": "Guerrero", "3": "Arquero", "4": "Paladín"}
        print(c("\n  Elige tipo:", Color.MORADO))
        for k, v in tipos.items():
            print(f"    {k}. {v}")
        while True:
            tipo_key = input(c("  Tipo (1-4): ", Color.MORADO)).strip()
            if tipo_key in tipos:
                break
            print(c("  Elige entre 1, 2, 3 o 4.", Color.ROJO))

        self.heroe_nombre = nombre
        self.heroe_tipo   = tipos[tipo_key]
        self.heroe_vida   = random.randint(80, 120)
        self.heroe_fuerza = random.randint(50, 100)

        # Mostrar el código que se "generó"
        print(c("\n  El código Python de tu objeto sería:", Color.AMARILLO))
        print(f"""
    mi_heroe = Heroe("{self.heroe_nombre}", {self.heroe_vida}, {self.heroe_fuerza})

    # Acceder a atributos:
    print(mi_heroe.nombre)   # → {self.heroe_nombre}
    print(mi_heroe.vida)     # → {self.heroe_vida}
    print(mi_heroe.fuerza)   # → {self.heroe_fuerza}

    # Llamar un método:
    mi_heroe.atacar()        # → ¡Ataca!
""")

        # Mostrar ficha del héroe
        separador()
        print(c(f"\n  ⚔  HÉROE CREADO: {self.heroe_nombre}", Color.NEGRITA))
        print(f"     Clase (tipo): {c(self.heroe_tipo, Color.MORADO)}")
        print(f"     Vida:         {c(str(self.heroe_vida), Color.VERDE)}")
        print(f"     Fuerza:       {c(str(self.heroe_fuerza), Color.ROJO)}")
        separador()

        self.score += 20
        print(c("\n  +20 puntos — ¡instanciaste tu primer objeto!", Color.VERDE))
        pausa()

    # ── RONDA 5: Quiz final ──────────────────

    def ronda5_quiz_final(self):
        self.ronda_actual = 5
        titulo("RONDA 5 — Quiz Final (Boss)")
        self.mostrar_progreso()

        print(c("  Última ronda. Sin pistas. ¡Tú puedes!\n", Color.AMARILLO))

        preguntas = [
            (
                "¿Cuál es la sintaxis CORRECTA para crear un objeto?",
                [
                    'heroe1 = new Heroe("Ana", 100)',
                    'heroe1 = Heroe("Ana", 100)',
                    'class heroe1 = Heroe("Ana")',
                    'Heroe.crear("Ana", 100)',
                ],
                1
            ),
            (
                "¿Qué representa 'self' en los métodos de una clase?",
                [
                    "El nombre de la clase",
                    "El objeto que llama al método",
                    "Un parámetro opcional de Python",
                    "La clase padre (herencia)",
                ],
                1
            ),
            (
                "¿Puedo crear múltiples objetos desde la misma clase?",
                [
                    "No, solo se puede crear un objeto por clase",
                    "Sí, y cada objeto puede tener sus propios valores",
                    "Sí, pero todos compartirán los mismos valores",
                    "Depende de si la clase tiene __init__",
                ],
                1
            ),
            (
                "¿Cuál de estos es un MÉTODO y no un atributo?",
                ["self.nombre", "self.vida", "def atacar(self):", "self.fuerza = 80"],
                2
            ),
            (
                "¿Cuál es la diferencia entre clase y objeto?",
                [
                    "No hay diferencia, son lo mismo",
                    "La clase es la instancia, el objeto es el molde",
                    "La clase es el molde, el objeto es la instancia creada",
                    "Un objeto solo puede tener atributos, no métodos",
                ],
                2
            ),
        ]

        pts_ronda = 0
        por_pregunta = 20 // len(preguntas)

        for i, (preg, ops, correcta) in enumerate(preguntas, 1):
            print(c(f"\n  Pregunta {i}/{len(preguntas)}", Color.AZUL))
            ok = self.opcion_multiple(preg, ops, correcta)
            if ok:
                pts_ronda += por_pregunta

        self.score = min(self.score + pts_ronda, 100)
        print(c(f"\n  +{pts_ronda} puntos en el quiz final", Color.AMARILLO))
        pausa()

    # ── RESULTADOS FINALES ───────────────────

    def resultados(self):
        titulo("RESULTADOS FINALES")

        s = self.score

        if s >= 90:
            medalla = c("🥇 MAESTRO DE POO", Color.AMARILLO)
            mensaje = "¡Eres un crack! Claramente entiendes clases y objetos."
        elif s >= 70:
            medalla = c("🥈 DESARROLLADOR PLATA", Color.AZUL)
            mensaje = "¡Muy bien! Repasa los conceptos que fallaste y lo clavás."
        elif s >= 50:
            medalla = c("🥉 APRENDIZ BRONCE", Color.VERDE)
            mensaje = "Vas por buen camino. Lee el resumen de abajo y vuelve a intentar."
        else:
            medalla = c("📚 MODO ESTUDIO", Color.MORADO)
            mensaje = "No te rindas. Lee el resumen, practica y vas a mejorar."

        print(f"""
  {medalla}
  Puntaje final: {c(str(s) + ' / 100', Color.NEGRITA)}
  {mensaje}
""")

        if self.heroe_nombre:
            print(c("  Tu héroe graduado:", Color.AMARILLO))
            print(f"  ⚔  {self.heroe_nombre} el {self.heroe_tipo} — {self.heroe_vida} vida, {self.heroe_fuerza} fuerza")
            print(f"  Código: {c(f'mi_heroe = Heroe(\"{self.heroe_nombre}\", {self.heroe_vida}, {self.heroe_fuerza})', Color.MORADO)}\n")

        separador()
        print(c("  RESUMEN: Lo que debes saber para el examen", Color.NEGRITA))
        print("""
  CLASE     → Plantilla / molde que define atributos y métodos
              Ejemplo: class Heroe:

  OBJETO    → Instancia creada desde la clase con datos propios
              Ejemplo: heroe1 = Heroe("Ana", 100, 85)

  ATRIBUTO  → Variable que pertenece al objeto (self.nombre)
              Guarda características del objeto

  MÉTODO    → Función definida dentro de la clase (def atacar(self):)
              Define las acciones que puede hacer el objeto

  SELF      → Referencia al objeto actual. Siempre es el primer
              parámetro de los métodos de instancia

  __INIT__  → Método constructor. Se ejecuta automáticamente
              al crear el objeto (inicializa atributos)

  INSTANCIAR→ Crear un objeto a partir de una clase:
              obj = MiClase(arg1, arg2)
""")
        separador()
        print(c("\n  ¡Suerte en el examen, Mateo! 🎓\n", Color.VERDE))

    # ── MÉTODO PRINCIPAL: jugar ──────────────

    def jugar(self):
        """Método que orquesta todo el juego."""
        self.intro()
        self.ronda1_concepto()
        self.ronda2_atributos_metodos()
        self.ronda3_leer_codigo()
        self.ronda4_instanciar()
        self.ronda5_quiz_final()
        self.resultados()

        # ¿Jugar de nuevo?
        resp = input(c("  ¿Jugar de nuevo? (s/n): ", Color.AZUL)).strip().lower()
        if resp == "s":
            self.__init__()  # reiniciar atributos
            self.jugar()


# ─────────────────────────────────────────
#  PUNTO DE ENTRADA
# ─────────────────────────────────────────
if __name__ == "__main__":
    juego = JuegoPOO()   # ← ¡Aquí instanciamos un objeto!
    juego.jugar()        # ← ¡Aquí llamamos un método!
```
